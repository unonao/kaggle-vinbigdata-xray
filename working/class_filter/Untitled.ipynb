```python
def get_img(path, use_meta=False ,ch1=False, ch2=False):
    """
    pathからimageの配列を得る
    """
    im_three = cv2.imread(path)
    print(im_three.shape)
    if use_meta:
        im_three[:,:,1] = ch1*255
        im_three[:,:,2] = ch2*255
    return im_three
```
